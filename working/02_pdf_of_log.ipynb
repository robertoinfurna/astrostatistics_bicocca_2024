```python
import numpy as np
import pylab as plt
plt.rcParams['figure.figsize'] = [8, 8]
```

```python
x = np.random.uniform(0.1,10,10000)
```

```python
n_bin = 100
plt.hist(x,n_bin,density=True,histtype='step',lw=2);
plt.xlabel("$x$");
```

```python
y = np.log10(x)
```

If $x$ is a random variable then $Log(x)$ is also a random variable.

To transform probability distributions when taking functions of random variables, we can simply use conservation of dimensionless probability, i.e. 

$$\mathrm{Prob}(x, x+dx) = \mathrm{Prob}(y, y+dy)$$

$$p(x)dx = p(y)dy.$$ 

where $y = \log(x)$.

Thus, $$p(y) = \left|\frac{dx}{dy}\right| p(x) = 10^y \ln(10) p (x)$$

```python
n_bin = 100
plt.hist(y,n_bin,density=True,histtype='step',lw=2);
plt.xlabel("$Log(x)$");

ygrid=np.linspace(-1,1,100)
p_x = 1/(10-0.1)   
f_y = np.log(10) * 10**ygrid * p_x

plt.plot(ygrid,f_y);

```
