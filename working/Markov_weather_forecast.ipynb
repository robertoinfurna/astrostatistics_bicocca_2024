```python
import numpy as np
from scipy import stats
from matplotlib import pyplot as plt
import numpy.random as random
```

```python
#clear = 0
#cloudy = 1

N = 2000

days = [0] #initialize weather as clear

sunny_days = 0 #initialize cumulative function
cumulative = []
for i in range(N): 
    if days[i]==0: 
        sunny_days = sunny_days+1
        days.append(random.choice([0,1], p=[0.9,0.1]))
    else:
        days.append(random.choice([0,1], p=[0.5,0.5]))
    cumulative.append(sunny_days/(i+1)) #cumulative.append(sunny_days/N)

plt.subplot(1, 2, 1)
plt.plot(cumulative)
plt.title("Trace Plot")
plt.xlabel("Number of days N")
plt.ylabel("Number of clear days over N")

plt.subplot(1, 2, 2)
plt.bar(["clear","cloudy"],[sunny_days/N,1-sunny_days/N], color=["orange","blue"])
plt.title("Distribution clear days")
plt.xlabel("Weather")

plt.show()


```
