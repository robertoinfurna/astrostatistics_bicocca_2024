```python
import numpy as np
from scipy import stats
from matplotlib import pyplot as plt
import random
import time
```

```python
def volume_hypersphere(N,M): #N number dimensions, M number of trials
    
    x = []
    for i in range(N): x.append(stats.uniform(0,1).rvs(M))
    x = np.array(x)
    
    sum = 0
    for i in range(N): sum = sum + x[i,:]**2
    r = np.sqrt(sum)
    
    return 2**N*np.sum(r < 1)/M

M = 1000 #number of random points
N=np.arange(2,1000,1)
V=[]
elapsed_time=[]

for n in N:
    t_start = time.process_time() 
    v = volume_hypersphere(n,M)
    t_stop = time.process_time()
    V.append(v)
    elapsed_time.append(t_stop-t_start)


plt.plot(N,v)
plt.show()
```

```python
plt.plot(N,elapsed_time)
plt.show()
```
